```python
import got
import csv
import pandas as pd
import re
```

```python
def makeQuery(term, maxTweets=500, user=''):
    criteria = got.manager.TweetCriteria()
    criteria.setSince("2018-01-01")
    criteria.setUntil("2018-10-28")
    criteria.setQuerySearch(term)
    criteria.setMaxTweets(maxTweets)
    if user!='':
        criteria.setUsername(user)
    t = got.manager.TweetManager.getTweets(criteria)
    return list(filter(lambda x: x.language == 'pt', t))
```

```python
bolsonaroQuery = makeQuery("bolsonaro")
bolsonaroList = [tweet.text.encode('utf-8') for tweet in bolsonaroQuery]

#haddadQuery = makeQuery("haddad")
#haddadList = [tweet.text for tweet in haddadQuery]
```

```python
with open('tweets.csv','wb') as file:
    for item in bolsonaroList:
        file.write(item + ";")
        file.write("\n")
```

```python
# HADDAD
print('\n-----------------------\nHADDAD TWEETS\n')
haddadTweets = makeQuery('haddad')
users = map(lambda x: x.username, haddadTweets)
haddadHistoryTweets = pd.DataFrame(columns=['user','tweet','date','favorites','retweets'])
print(haddadTweets)
c = 0
for user in users:
    print('\n---------------------\nuser %i of %i' % (c,len(users)))
    c+=1
    tmp = makeQuery('haddad',maxTweets=200,user=user)
    for i in tmp:
        haddadHistoryTweets = haddadHistoryTweets.append({
            'user': i.username,
            'tweet': i.text,
            'date': i.date,
            'favorites': i.favorites,
            'retweets': i.retweets
        }, ignore_index=True)

display(haddadHistoryTweets.head())    
```

```python
def altNameQuery(altName):
    print('\n-----------------------\n%s TWEETS\n'%altName)
    bozoAltTweets = makeQuery(altName,maxTweets=34)
    users = map(lambda x: x.username, bozoAltTweets)
    bozoAltHistoryTweets = pd.DataFrame(columns=['user','tweet','date','favorites','retweets'])
    c = 0
    for user in users:
        print('\n---------------------\nuser %i of %i' % (c,len(users)))
        c+=1
        tmp = makeQuery(altName,maxTweets=200,user=user)
        for i in tmp:
            bozoAltHistoryTweets = bozoAltHistoryTweets.append({
                'user': i.username,
                'tweet': i.text,
                'date': i.date,
                'favorites': i.favorites,
                'retweets': i.retweets
            }, ignore_index=True)

    return bozoAltHistoryTweets
```

```python
# BOZO ALT NAMES
print('\n-----------------------\nBOZO ALT TWEETS\n')
altNames = ['bozonaro','salnorabo','bonoro']
bozoAltHistoryTweets = pd.DataFrame(columns=['user','tweet','date','favorites','retweets'])
for name in altNames:
    bozoAltHistoryTweets = bozoAltHistoryTweets.append(altNameQuery(name))
    
display(bozoAltHistoryTweets.head())
```

```python
#remove tweets that contains bolsonaro and haddad comments together
cleanBozoTweets = bozoHistoryTweets[bozoHistoryTweets.tweet.str.contains('lula|haddad|pt')]
cleanHaddadTweets = haddadHistoryTweets[haddadHistoryTweets.tweet.str.contains('bolsonaro')]

def removeLink(string):
    return re.sub(r'http\S+', '', string)

#removendo links
cleanBozoTweets['tweet'] = cleanBozoTweets['tweet'].map(lambda x: removeLink(x))
cleanHaddadTweets['tweet'] = cleanHaddadTweets['tweet'].map(lambda x: removeLink(x))
print(cleanBozoTweets)

cleanHaddadTweets.to_csv('haddad.csv',sep=',',encoding='utf-8')
cleanBozoTweets.to_csv('bozo.csv',sep=',',encoding='utf-8')
bozoAltHistoryTweets.to_csv('bozoAlt.csv',sep=',',encoding='utf-8')
```
